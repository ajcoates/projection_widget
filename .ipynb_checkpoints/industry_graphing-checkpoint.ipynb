```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Bar chart
# Pie chart
# Group 'cnt' into ranges, use clustered bar chart to show similarities between industries
# Shape = (51,2)
df = pd.read_csv('iheart_industries.csv')
df.sort_values(by='cnt', ascending = False, inplace = True)
df1 = df.iloc[0:17]
df2 = df.iloc[18:35]
df3 = df.iloc[36:]
df_top_10 = df.iloc[1:11]
```

```python
df3.shape
```

```python
x = df_top_10['industry']
y = df_top_10['cnt']
low = min(y)
high = max(y)
plt.ylim([math.ceil(low-0.5*(high-low)), math.ceil(high+0.5*(high-low))])
plt.xticks(rotation='vertical')
plt.bar(x,y) 
plt.show()
```

```python
plt.clf()

# using some dummy data for this example
xs = df_top_10['industry']
ys = df_top_10['cnt']

plt.bar(xs,ys)

# zip joins x and y coordinates in pairs
for x,y in zip(xs,ys):

    label = int(float("{:.2f}".format(y)))

    plt.annotate(label, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center

plt.xticks(np.arange(0,10,1), rotation = -45, ha='left')
# plt.ylim([math.ceil(low-0.5*(high-low)), math.ceil(high+0.5*(high-low))])
plt.ylim(0,2100)
plt.ylabel('Project Count')
# plt.title("iHeart's Top 10 Industry Uses for AdWire, 01/01/2020 - 09/16/2020")
plt.title("iHeart's Top 10 Industry Uses for AdWire, 01/01/2018 - 11/01/2020 ('Business/Organization' only)")


plt.show()
```
