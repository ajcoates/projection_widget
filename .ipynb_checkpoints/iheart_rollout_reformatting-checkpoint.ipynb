```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('The Rolling Out Stones Sep 2020 .csv')
users = pd.read_csv('users_20201101.csv')
```

```python
df = df.rename(columns = {'sea':'name'})
```

```python
# new = df["name"].str.split(" ", n = -1, expand = True)
new = df['name'].str.rsplit(' ', 1, expand = True)
new = new.rename(columns = {0: 'first', 1: 'last'})
df = pd.concat([df, new], axis=1)
df.shape
```

```python
df = df[['name', 'first', 'last', 'Email', 'Job Title', 'Phone', 'Mobile', '1-2-1',
       'Division President', 'Community Area President', 'Market President',
       'Area', 'Market', 'Credit Caps', 'Testimonial Video',
       'Testimonial Video (Cont.)', 'WIN', 'Testimonials', 'Count',
       'First Project', 'Second Project', 'Third Project', 'Fourth Project',
       'Fifth Project', 'Sixth Project', 'Seventh Project', 'Eighth Project',
       'Ninth Project', 'Tenth Project', 'Eleventh Project', 'Twelfth Project',
       'Thirteenth Project', 'Fourteenth Project', 'Fifteenth Project',
       'Sixteenth Project', 'Seventeeth Project', 'Eighteenth Project',
       'Nineteenth Project', 'Twentieth Project', 'Twenty First Project',
       'Twenty Second Project', 'Twenty Third Project', 'Run a Project?',
       'Onboarded', 'Scheduled', 'Missed/NA', 'Count of AEs', 'Ran 1',
       'Ran at least 2', 'Running 0', 'Total AEs', 'Notes on non-usage']]

df = df.rename(columns = {'name': 'Name', 'first': 'First', 'last':'Last'})

# df.drop(['Name'], axis = 1, inplace = True)
```

```python
# strips non numeric characters from Phone
df['Phone'] = df['Phone'].str.replace(r'\D+', '')

# converts empty values to NaN
df = df.replace(r'^\s*$', np.nan, regex=True)

# strips non numeric characters from Mobile
df['Mobile'] = df['Mobile'].str.replace(r'\D+', '')

# converts empty values to NaN
df = df.replace(r'^\s*$', np.nan, regex=True)

df['Phone'] = df['Phone'].fillna(1111111111)
df['Mobile'] = df['Mobile'].fillna(1111111111)
# df.head(50)
```

```python
# converts 1111111111 formatted numbers to 111-111-1111 for Phone and Mobile columbs
df['Phone']=df['Phone'].astype(str).apply(lambda x: ''+x[:3]+'-'+x[3:6]+'-'+x[6:10])
df['Mobile']=df['Mobile'].astype(str).apply(lambda x: ''+x[:3]+'-'+x[3:6]+'-'+x[6:10])

# converts 111-111-1111 (NaNs) back to NaN
df = df.replace('111-111-1111', np.nan)
```

```python
# df.to_csv('ros_sep2020_namesplit.csv', index=0)
```

```python
# Section below treats the df 'users', which includes project counts per employee
```

```python
# changes any value in Job Title containing "Account " (space included
# because it distinguishes 'accounting' from account execs, for example)
# to 'Account Executive'
df.loc[df['Job Title'].str.contains('Account ', na=False), 'Job Title'] = 'Account Executive'
```

```python
users.head()
```

```python
df_ae = df[df['Job Title'] == 'Account Executive']
users = users[users['name'] != 'David Liu']
users = users[users['name'] != 'James Palmiotto']
users = users[users['name'] != 'Dan Lankford']
users = users[users['name'] != 'Louis Wortelmann']
users = users[users['name'] != 'IHM Admin']
# users = users[len(str(users['name'])) < 2]

df_ae = df_ae.rename(columns = {'Name': 'name'})
```

```python
ae_list = list(df_ae['name'])
```

```python
users['t_f'] = users.name.isin(ae_list)
```

```python
users.t_f = users.t_f.astype(str)
users.loc[users.t_f.str.contains('True'), 't_f'] = 'AE'
users.loc[users.t_f.str.contains('False'), 't_f'] = 'Not Listed as AE'
```
